```python
base_dir = os.path.join(DESKTOP, "Concept_Detection")
```

```python
client = pymongo.MongoClient()
db = client.WATSON
```

```python
xls = pd.read_excel(os.path.join(base_dir, "corpus-02.xlsx"), sheetname="Mentions")
```

```python
ncit = pd.read_csv(os.path.join(DOCS_DIR, "Curation", "thesaurus", "Thesaurus.txt"), sep="\t", header=None)
```

```python
for (index, series) in xls.iterrows():
    nci_codes = str(series[3]).strip()
    if nci_codes != "" and nci_codes.startswith("C"):
        print nci_codes
```

```python
my_types = []
reviewed_types = []
for (index, series) in xls.iterrows():
    nci_codes = str(series[3]).strip()
    if nci_codes != "" and nci_codes.startswith("C"):
        nci_codes = nci_codes.split(",")
        #print nci_codes
        types = [ list(ncit[ncit[0] == nci_code.strip()].iloc[0])[-1] for nci_code in nci_codes ]
        my_types.append(", ".join(types))
    else: my_types.append("NA")
    reviewed_code = str(series[5])
    if reviewed_code.startswith("C"):
        type_ = list(ncit[ncit[0] == reviewed_code].iloc[0])[-1]
        reviewed_types.append(type_)
    else: reviewed_types.append("NA")
        
xls["NCI Semantic Type"] = pd.Series(my_types)
xls["NCI Semantic Type for Added Concepts"] = pd.Series(reviewed_types)
xls.to_excel(os.path.join(base_dir, "corpus-02_lucene_v1.xlsx"), index=False)
```
