```python
corpus = []
types = []
with open("/Users/singhv/Documents/Curation/thesaurus/owl/ThesaurusSemTypePrefName.tsv") as f:
    for line in f:
        line = line.strip().split("\t")
        if (len(line) > 2) and (line[-1].strip() != "null"):
            corpus.append(line[1].strip())
            types.append(line[2].strip())
```

```python
from sklearn.cross_validation import train_test_split
from itertools import groupby

X_train, X_test, y_train, y_test = train_test_split(corpus, types)

print "X_train size: %d" % len(X_train)
print "X_test size: %d" % len(X_test)
print "y_train size: %d" % len(y_train)
print "y_test size: %d" % len(y_test)
'''
all_prop = 0.0

for k, g in groupby(sorted(y_train)):
    prop = len(list(g))/float(len(y_train))
    all_prop += prop
    print "Proportion of %s is %f" % (k, prop)
    
print "Total prop is: %f" % all_prop

for k, g in groupby(sorted(y_test)):
    prop = len(list(g))/float(len(y_test))
    all_prop += prop
    print "Proportion of %s is %f" % (k, prop)
    
print "Total prop is: %f" % all_prop
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=1)
X = vectorizer.fit_transform(X_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline

#clf = MultinomialNB().fit(X, y_train)
clf = Pipeline([('vect', CountVectorizer()),
                ('tfidf', TfidfTransformer()),
                ('clf', SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, n_iter=5, random_state=42))])
_ = clf.fit(X_train, y_train)
```
