```python
import re, os, sys, difflib
from com.ibm.watson.datasource.clinicaltrials import ClinicalStudy
```

```python
nct_id = "NCT02278133"
cs = ClinicalStudy(nct_id)
cs.get()
raw_text = cs._xml.eligibility.criteria.textblock.strip()
print raw_text
```

```python
stripped_text = open("/Users/singhv/dev/scripts/temp/NCT02278133.txt").read().decode("utf-8")
print stripped_text
```

```python
seq_matcher = difflib.SequenceMatcher()
seq_matcher.set_seq1(raw_text)
seq_matcher.set_seq2(stripped_text)
for tag, i1, i2, j1, j2 in seq_matcher.get_opcodes():
    print ("%7s a[%d:%d] (%s) b[%d:%d] (%s)" % (tag, i1, i2, raw_text[i1:i2], j1, j2, stripped_text[j1:j2]))
```
